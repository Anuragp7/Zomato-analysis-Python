```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import os 
```

```python
os.chdir("C:\\Users\\DELL\\Downloads")
df =  pd.read_csv("Zomato data .csv")
df.head()
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

df['rate']=df['rate'].apply(handleRate)
df.head()
```

```python
df.info()
```

"now here we saw in summary there is no null values in it"

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("TYPE OF RESTURANT")
```

here we large no of dining resturant preferd by individual 

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker ="o")
plt.xlabel("Type of resturant", c="red",size=20)
plt.ylabel("Votes", c="red" ,size=20)
```

majority of resturant receive rating

```python
plt.hist(df['rate'],bins=5)
plt.title("Rating Distribution")
plt.show
```

The majority of resturant receive rating from 3.5 to 4

The majority of couple prefer resturants with an approximate cost of 300 rupees.

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data =df)
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order' , aggfunc='size',fill_value=0)
sns.heatmap(pivot_table , annot=True , cmap="YlGnBu" , fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("listed In (Type)")
plt.show()
                           
```
